```python
import numpy as np
import pandas as pd
# Making the data have the correct header for each column
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('C:\\Users\\Samuel\\python3stuff\\imports-85.data',names=cols)
cars.head()
```

```python
cars.columns #seeing that our reading of the data worked.
```

```python
#https://archive.ics.uci.edu/ml/datasets/automobile is the source of the data, and gives the dtypes of each column
#Since we are only using numeric data in our predictions, we can make a dataset that ony has int and float data
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
numeric_cars.replace('?',np.nan) #replacing the ? in the data with nan (not a number)
```

```python
#changing the dtypes in our numerical columns to floats, many were objects bc of the ?s in them
numeric_cars = numeric_cars.astype(float) 
#seeing how many nan values there are in each column. KNeighbors doesn't except them so we have to figure out how to change them
numeric_cars.isnull().sum()
```

```python
#We dropped the all the rows that have nan values in the price column bc there were only four and price is our prediction column
numeric_cars=numeric_cars.dropna(subset=['price'])
numeric_cars.isnull().sum() # seeing how many nan values are left
```

```python
#for the rest of the columns we replace the nan values with the mean of their columns
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum() #seeing how many nan values are left, should be 0
```

```python
#Here we are normalizing the data to make every entry a value from 0 to 1. We do this for every column, except the price column
#This is so our predictions are in a formatt we understand
price_col = numeric_cars['price']
numeric_cars = (numeric_cars.max() - numeric_cars)/(numeric_cars.max())
numeric_cars['price'] = price_col
numeric_cars.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

#creating a function that takes one column and the default k value,5, to make find our error metric, rsme, on price predictions

def knn_train_test(test_col,target_col, df_object):
    knn = KNeighborsRegressor(algorithm = 'brute') #instaniating and instance of KNeihborsRegressor class
    np.random.seed(1) #setting a random seed so the shuffle will be the same on all functions
    shuffle_index = np.random.permutation(df_object.index) #shuffling the data to eliminate data bias
    shuffled_df = df_object.reindex(shuffle_index) #reindexing the data
    half_df = int(len(shuffled_df)/2) #getting the numerical value for half the number of rows of the dataset
    train_df = shuffled_df.iloc[:half_df] #setting our training dataset to the first half of the full dataset
    test_df = shuffled_df.iloc[half_df:] #setting our test dataset to the second half of the full dataset
    knn.fit(train_df[[test_col]],train_df[target_col]) # fitting the dataset
    predictions = knn.predict(test_df[[test_col]]) #making the predictions on the price column of the test dataset
    mse = mean_squared_error(test_df[target_col],predictions) #fiding the mean squared error for predicted prices
    rmse = np.sqrt(mse) #finding the root mean squared error
    return rmse
rmse_vals = {} #creating a empty dictionary for the column and corresponding rsme value
numeric_cars_col = numeric_cars.columns.drop('price') #making a dataset with all numercal columns except our target column

#creating a loop that passes each column into the function and passes the column name and the rsme inoto rsme_vals dict
for col in numeric_cars_col:
    rmse_val = knn_train_test(col,'price',numeric_cars)
    rmse_vals[col]=rmse_val
    
    
    
# Create a Series object from the dictionary so 
# we can easily view the results, sort, etc
rmse_results_series = pd.Series(rmse_vals)
rmse_results_series.sort_values()
    

    
    
    
    
```

```python
#function that has mutiple columns and multiple k-values. K values range from 1-25
def knn_train_test(test_cols,target_col, df_object):
    
    np.random.seed(1)
    shuffle_index = np.random.permutation(df_object.index)
    shuffled_df = df_object.reindex(shuffle_index)
    half_df = int(len(shuffled_df)/2)
    train_df = shuffled_df.iloc[:half_df]
    test_df = shuffled_df.iloc[half_df:]
    k_vals =[i for i in range(1,25)]
    k_rmses={}
    for k in k_vals:
        knn = KNeighborsRegressor(n_neighbors=k,algorithm = 'brute')
        knn.fit(train_df[test_cols],train_df[target_col])
        predictions = knn.predict(test_df[test_cols])        
        mse = mean_squared_error(test_df[target_col],predictions)
        rmse = np.sqrt(mse)
        k_rmses[k] = rmse
    return k_rmses
k_rsme_results_two={} #dictionary to store the results

two_best_features = knn_train_test(['horsepower','curb-weight'],'price', numeric_cars)
k_rsme_results_two['two best features'] = two_best_features

three_best_features =knn_train_test(['horsepower','curb-weight','width'],'price', numeric_cars)
k_rsme_results_two['three best features'] = three_best_features

four_best_features =knn_train_test(['horsepower','curb-weight','width','city-mpg'],'price', numeric_cars)
k_rsme_results_two['four best features'] = four_best_features


    
k_rsme_results_two
   
        
    
```
